### Separate the dataset into Predictors and Target
Diagnosis attribute is the Target and the rest of the features are Predictors.

```python
X= data.iloc[:,2:32]
X.shape
```

```python
X.head()
```

```python
y= data.diagnosis
y.shape
```

```python
y.head()
```

##### Converting categorical values in diagnosis to numerical values

```python
y_num= pd.get_dummies(y)
y_num.tail(10)
```

```python
y=y_num.M # We are considering only the Maliignant column
y.tail()
```

```python
y.shape
```

```python
plt.figure(figsize=(20,10))
sns.heatmap(X.corr(),annot=True,linewidths=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=20)
X_train.shape
```

#### Normalizing the training data

```python
from sklearn.preprocessing import MinMaxScaler
scaler= MinMaxScaler()
```

```python
X_train_normalised= scaler.fit_transform(X_train)
X_train_normalised
```

```python
X_test_normalised= scaler.fit_transform(X_test)
X_test_normalised
```

### Training using SVM

```python
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, classification_report
```

```python
model = SVC()
```

```python
model.fit(X_train_normalised, y_train)
```

```python
y_pred= model.predict(X_test_normalised)
```

```python
cm= confusion_matrix(y_test, y_pred)
cm
```

```python
print(classification_report(y_test, y_pred))
```

### The model has achieved an accuracy of 96%
